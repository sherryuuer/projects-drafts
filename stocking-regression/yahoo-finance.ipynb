```python
# import packages
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

%matplotlib inline
# define function to patch process
days = 20
# company: ticker
tickers = {
    'apple': 'AAPL',
    'microsoft': 'MSFT',
    'google': 'GOOGL',
    'facebook': 'META',
    'amazon': 'AMZN'
}


def get_data(tickers, days):
    df = pd.DataFrame()
    for company in tickers:
        print(company, 'is processing...')
        ticker = tickers[company]
        tkr = yf.Ticker(ticker)
        hist = tkr.history(period=f'{days}d')
        hist.index = hist.index.strftime('%d %B %Y')
        hist = hist[['Close']]
        hist.columns = [company]
        hist = hist.T
        hist.index.name = company
        df = pd.concat([df, hist])
    return df

df = get_data(tickers, days)
df.head()
```

```python
import altair as alt

# df.loc[tickers.keys()]
df.sort_index()
df_t = df.T.reset_index()
df_t.head()
```

```python
df_melt = pd.melt(df_t, id_vars=['Date']).rename(columns={
    'variable': 'company',
    'value': 'Stock Prices(USD)'})
df_melt.head()
```

```python
chart = (
    alt.Chart(df_melt).mark_line(opacity=0.8, clip=True)
    .encode(
        x="Date:T",
        y=alt.Y("Stock Prices(USD):Q", stack=None),
        color="company:N"
    )
)
```
