### 1. Data preparation and preprocessing

data: internet movie database

```python
# import libraries
import pandas as pd
import numpy as np
import scipy.stats as st
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("imdb.csv")
data.head()
```

```python
# Pick some variables
df = data[["IMDB_Rating", "Released_Year", "Runtime",
           "Meta_score", "Gross", "Director"]]
df.head()
```

```python
# get the info from the data
df.info()
```

```python
# turn "Released_Year" to numeric data type if can't turn it into NaN
df.loc[:, "Released_Year"] = pd.to_numeric(df["Released_Year"],
                                    errors="coerce")
```

```python
# Fill NaN values with a default value (e.g., 0)
df["Released_Year"].fillna(0, inplace=True)

# Perform floor division and multiplication
df.loc[:, "Decade"] = np.floor(df["Released_Year"] / 10) * 10
```

```python
df.head()
```

```python
# convert and transform Runtime
df.loc[:, "Runtime"] = df["Runtime"].str.replace(" min", "").astype(int)
```

```python
# fill NaN value with 0
df["Gross"].fillna(0, inplace=True)
df["Gross"].isna().sum()
```

```python
df.loc[:, "Gross"] = df["Gross"].str.replace(",", "").fillna(0).astype(int)
```

```python
df["Gross_scaled"] = df["Gross"] / 1000000
```

```python
# drop columns
df = df.drop(columns = ["Released_Year", "Gross"])
```

```python
df.head()
```

```python
df["Gross_scaled"] = df["Gross_scaled"].astype(float)
```

```python
df["Runtime"] = df["Runtime"].astype(int)
```

### 2. Exloratory data analysis

```python
# summary statistics
df.describe()
```

```python
# plot the data "IMDB_Rating", kde means Kernel Density Estimation.
sns.histplot(df["IMDB_Rating"], kde=True)
plt.title("IMDB Rating Distribution")
plt.show()
```

```python
# get the top 10 directors with the highest avg IMDB rating
director_ratings = df.groupby("Director")["IMDB_Rating"].mean().sort_values(ascending=False).head(10)
director_ratings
```

```python
# Visualize the Mean IMDB ratings for the top 10 directors
plt.figure(figsize=(7, 5))
sns.barplot(x = director_ratings.index,
            y = director_ratings.values)
plt.xticks(rotation=45)
plt.xlabel('Director')
plt.ylabel('IMDB average rating')
plt.show()
```

```python
# Distribution of IMDB Rating by decade
plt.figure(figsize=(12, 4))
sns.boxplot(x="Decade", y="IMDB_Rating", data=df)
plt.title("IMDB Distribution by decade")
plt.show
```

```python
# Covariance matrix
df_co = df[['Runtime', 'IMDB_Rating', 'Gross_scaled', 'Meta_score']]
df_co.cov()
```

```python
# Correlation heatmap
plt.figure(figsize=(7, 5))
sns.heatmap(df_co.corr(), annot=True, cmap="coolwarm")
plt.show
```

```python
# Calculate the Mean, the standard deviation, the sample size
summary = df.groupby("Decade").agg({"Gross_scaled": ["mean", "std", "count"]})
summary.columns = ["Gross_mean", "Gross_std", "Gross_count"]
summary["Gross_sem"] = summary["Gross_std"] / np.sqrt(summary["Gross_count"])
summary
```

```python
# Calculate the confidence interval with 95% cl
alpha = 0.05
summary["tscore"] = st.t.ppf(1 - alpha / 2, summary["Gross_count"] - 1)

# confidence interval
summary["lower_ci"] = summary["Gross_mean"] - summary["tscore"] * summary["Gross_sem"]
summary["upper_ci"] = summary["Gross_mean"] + summary["tscore"] * summary["Gross_sem"]

summary
```
